```python
def get_product(x:int,y:int):
    return x*y
```

```python
get_product(x=2,y="3")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_producte(x:int, y:int):
    return x*y
```

```python
get_producte(x=2,y=3)
# 6
```

```python
get_producte(x=2,y="3")
# error because of ensure_annotations type of arguments
```
